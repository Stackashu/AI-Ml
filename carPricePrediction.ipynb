```python
x_one_encode = pd.get_dummies(
    X , columns=['model','transmission','fuelType',],drop_first=True
)
```

```python
x_one_encode = x_one_encode.astype(int)
```

```python
x_one_encode
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

Xlable = X.copy()  # make a safe copy
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    Xlable[col] = le.fit_transform(Xlable[col].astype(str))  # Convert to string in case of nulls
    label_encoders[col] = le
```

```python
Xlable
```

```python
#here we are just standardizing the values to keep them small it can be negative too but don't effect
from sklearn.preprocessing import StandardScaler
cols = ['year','tax','mileage','mpg']
scaler = StandardScaler()
x_one_encode[cols] = scaler.fit_transform(x_one_encode[cols])
```

```python
x_one_encode
```

```python
# import pandas as pd
from scipy.stats import pearsonr

# Suppose your dataset looks like this
# X = features dataframe
# y = target variable (Series)

selected_features = []
correlation_results = {}

for col in x_one_encode.columns:
    # Only test numeric columns (pearsonr requires numbers)
    if pd.api.types.is_numeric_dtype(x_one_encode[col]):
        corr, p_value = pearsonr(x_one_encode[col], y)
        
        correlation_results[col] = {'correlation': corr, 'p_value': p_value}
        
        # You can set conditions, e.g. |corr| > 0.2 and p < 0.05
        if abs(corr) > 0.2 and p_value < 0.05:
            selected_features.append(col)

# Convert results to DataFrame for better view
correlation_df = pd.DataFrame(correlation_results).T
print("Correlation Results:\n", correlation_df)

print("\nSelected Features based on Pearson r:\n", selected_features)
```

```python
# Your selected features
selected_features = [
    'year', 'mileage', 'tax', 'mpg', 'engineSize',
    'model_ Edge', 'model_ Fiesta', 'model_ Kuga', 'model_ Mustang',
    'transmission_Manual', 'fuelType_Petrol'
]

# Keep only these in X
X_selected = x_one_encode[selected_features]

print("Shape of new feature set:", X_selected.shape)
X_selected.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.2, random_state=42)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Score
print("Train RÂ²:", model.score(X_train, y_train))
print("Test RÂ²:", model.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Make predictions on the test set
y_pred = model.predict(X_test)

# Scatter plot
plt.figure(figsize=(7,5))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)

# Add a diagonal reference line (perfect predictions)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--')

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Predicted vs Actual Prices")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
print("RF Test RÂ²:", rf.score(X_test, y_test))

gb = GradientBoostingRegressor(random_state=42)
gb.fit(X_train, y_train)
print("GB Test RÂ²:", gb.score(X_test, y_test))
```

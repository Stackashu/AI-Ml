```python
def episode_extractor(txt):
    brk = False
    epes = ""
    for i in txt:
        if i == ")":
            return epes
        if brk:
            epes += i
        if i == "(":
            brk = True
    return epes  
```

```python
df['Episodes']=df["Title"].apply(episode_extractor)
```

```python
def extract_timeStamps(txt):
    check = False
    count = 19
    text = ""
    for i in txt:
        if count == 0:
            return text
        if i == ')':
           check = True
           continue
        if check == True:
            count -= 1
            text += i
        
```

```python
df['TimeStamps'] = df['Title'].apply(extract_timeStamps)
```

```python
from datetime import datetime

def count_months_in_range(date_range_str):
    # Example: "Apr 2009 - Jul 2010"
    start_str, end_str = date_range_str.split(" - ")
    
    # Convert to datetime (day is set to 1 by default)
    start_date = datetime.strptime(start_str, "%b %Y")
    end_date = datetime.strptime(end_str, "%b %Y")
    
    # Calculate total months
    months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1
    return months
```

```python
df["Total Month"] = df["TimeStamps"].apply(count_months_in_range)
```

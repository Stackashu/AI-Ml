```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('heart.csv')
```

```python
df['HeartDisease'].value_counts().plot(kind='bar')
```

```python
def plotGraph(var,num): 
    plt.subplot(2,2,num)
    sns.histplot(df[var],kde=True)

plotGraph('Age',1)
plotGraph('RestingBP',2)
plotGraph('Cholesterol',3)
plotGraph('MaxHR',4)

plt.tight_layout()
```

```python
chol_Mean = df.loc[df['Cholesterol'] !=0 ,'Cholesterol'].mean() 
chol_Mean
```

```python
df['Cholesterol'] = df['Cholesterol'].replace(0,chol_Mean)
```

```python
df['Cholesterol'].value_counts()
```

```python
sns.histplot(df['Cholesterol'],kde=True)
```

```python
resting_Mean = df.loc[df['RestingBP'] != 0 , 'RestingBP'].mean() 
```

```python
df['RestingBP'] = df['RestingBP'].replace(0,resting_Mean)
```

```python
sns.histplot(df['RestingBP'],kde=True)
```

```python
def plotGraph(var,num): 
    plt.subplot(2,2,num)
    sns.histplot(df[var],kde=True)

plotGraph('Age',1)
plotGraph('RestingBP',2)
plotGraph('Cholesterol',3)
plotGraph('MaxHR',4)

plt.tight_layout()
```
